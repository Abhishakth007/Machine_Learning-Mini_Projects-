```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("C:/Users/Abhi/Downloads/test_scores.csv")
df = pd.DataFrame(data)
```

```python
math_array = np.array(df['math'])
cs_array = np.array(df['cs'])
print(math_array)
print(cs_array)
```

```python
iterations = 100000
n= len(cs_array)
m_current = 0
b_current = 0
learning_rate = 0.00021
for i in range(iterations):
    y_predicted = m_current*math_array + b_current
    sum_of_least_squares = 1/n * sum([val*val for val in cs_array - y_predicted])
    derivative_of_m = -(2/n) * sum(math_array * (cs_array - y_predicted) )
    derivative_of_b = -(2/n) * sum(cs_array - y_predicted )
    m_current = m_current - learning_rate * derivative_of_m
    b_current = b_current - learning_rate * derivative_of_b
    print(f"Slope_m = {m_current} , Intercept_b = {b_current} , Cost_Function = {sum_of_least_squares}")
    


```

```python
l_model = linear_model.LinearRegression()
l_model.fit(df[['math']],df['cs'])
print(l_model.coef_)
print(l_model.intercept_)
```
