```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

from sklearn import linear_model
```

```python
data = pd.read_csv("C:/Users/Abhi/OneDrive/Desktop/area_vs_prices.csv")
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
plt.xlabel("Area(In Sq_Feet)")
plt.ylabel("Prices in Rupees")
plt.scatter(df["area"],df["prices"],marker = "+")
```

```python
l_model = linear_model.LinearRegression()
l_model.fit(df[['area']],df['prices'])  # x = independant variable represented by 2d array and y is dependant so 1d array
print(l_model.coef_)    # The slope value
print(l_model.intercept_)  # The Intecept or constant value
```

```python
predicted_price = l_model.predict([[330]])
print(predicted_price)
```
