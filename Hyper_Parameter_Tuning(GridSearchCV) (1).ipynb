```python
from sklearn import datasets
import pandas as pd
```

```python
dt = datasets.load_digits()
```

```python
df = pd.DataFrame(dt.data)
df.columns = dt.feature_names
df.head()
```

```python
target_df = pd.DataFrame(dt.target)
target_df.columns = ['target']

target_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df,target_df,test_size = 0.25)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.ensemble import RandomForestClassifier
```

```python
model_svc = SVC()   # Parameters = C  , Kernel , Gamma
model_svc.fit(X_train,y_train)
model_svc.score(X_test,y_test)
```

```python
model_dct = DecisionTreeClassifier() # Criterion = gini
model_dct.fit(X_train,y_train)
model_dct.score(X_test,y_test)
```

```python
model_l = LogisticRegression() # C = Regularization (small C = Strong Regularization)
model_l.fit(X_train,y_train)
model_l.score(X_test,y_test)
```

```python
model_gnb = GaussianNB()
model_gnb.fit(X_train,y_train)
model_gnb.score(X_test,y_test)
```

```python
model_mn = MultinomialNB()
model_mn.fit(X_train,y_train)
model_mn.score(X_test,y_test)
```

```python
model_rfc = RandomForestClassifier() # n_estimators = Divides into multiple trees
model_rfc.fit(X_train,y_train)
model_rfc.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# SVC Parameter Tuning
import numpy as np 

kernels = ['linear', 'poly', 'rbf', 'sigmoid']
C = [10,25,50]
average_scores_svc = {}

for k_val in kernels:
    for c_val in C:
        scores_array = cross_val_score(model_svc,df,target_df,cv = 5)
        mean_score = np.average(scores_array)
        average_scores_svc[k_val + '_' + str(c_val)] = mean_score
        
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'kernel' : ('linear', 'poly', 'rbf', 'sigmoid'),'C' : [10,25,50]}

gscv = GridSearchCV(model_svc,parameters,cv = 10,return_train_score = False)
gscv.fit(X_train,y_train)

gscv.cv_results_
```

```python
scores = []

final_data = {'Logisticregression' : {'model' : model_l, 'params' : { 'C' : [3,7,9]}
             },
             'SVM' : {'model' : model_svc,'params' : {'kernel' : ('linear', 'poly', 'rbf', 'sigmoid'),'C' : [10,25,50]}},
             'RandomForestClassifier' : {'model' : model_rfc,'params' : {'n_estimators' : [10,20,30,40,50]}}}
```

```python
for key,values in final_data.items():
    gscv = GridSearchCV(values['model'],values['params'],cv = 10,return_train_score = False)
    gscv.fit(X_train,y_train)
    scores.append({'model' :values['model'] ,
                  'best_score' : gscv.best_score_,
                  'best_params' : gscv.best_params_})
```

```python
ultimate_df = pd.DataFrame(scores)
ultimate_df
```
